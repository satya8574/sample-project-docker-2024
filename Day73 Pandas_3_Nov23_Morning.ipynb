```python
data = movies.merge(directors, how = "left", left_on = "director_id", right_on = "id")
```

```python
data.head(10)
```

```python
data.drop(["director_id", "id_y"], axis = 1, inplace = True)
```

```python
data.head()
```

```python
def encode_gender(x):
  if x == "Male":
    return 0
  else:
    return 1
```

```python
data["gender"] = data["gender"].apply(encode_gender)
```

```python
data
```

```python
data[["revenue", "budget"]].apply(np.mean)
```

```python
def profit(row):
  return row["revenue"] - row["budget"]
```

```python
data[["revenue", "budget"]]
```

```python
data["profit"] = data[["revenue", "budget"]].apply(profit, axis = 1)
```

```python
# Create a new column called risky which has a value True for movies
# where the budget is greater than the average revenue of movies by the director

def func(x):
  x["risky"] = (x["budget"] - x["revenue"].mean()) >= 0
  return x

data_risky = data.groupby("director_name", group_keys = False).apply(func)
```
