```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("movies.csv", index_col = 0)
directors = pd.read_csv("directors.csv", index_col = 0)
```

```python
data = movies.merge(directors, how = "left", left_on = "director_id", right_on = "id")
```

```python
data.drop(["director_id", "id_y"], axis = 1, inplace = True)
```

```python
# FIND OUT THE MOST PRODUCTICE DIRECTOR!
```

```python
data.groupby("director_name")["title"].count().sort_values(ascending = False)
```

```python
data_agg = data.groupby("director_name")[["year", "title"]].aggregate({
    "year": ["min", "max"],
    "title": "count"
})
```

```python
data_agg
```

```python
data_agg.columns
```

```python
data_agg["year"]["min"]
```

```python
data_agg["title"]
```

```python
data_agg = data.groupby("director_name")[["year", "title"]].aggregate(
    year_max = ("year", "max"), # new_column = ("old_column", "agg")
    year_min = ("year", "min"),
    title_count = ("title", "count")
)
```

```python
data_agg
```

```python
data_agg.reset_index()
```

```python
data_agg["years_active"] = data_agg["year_max"] - data_agg["year_min"]
```

```python
data_agg
```

```python
data_agg["movies_per_year"] = data_agg["title_count"] / data_agg["years_active"]
```

```python
data_agg
```

```python
data_agg.sort_values("movies_per_year", ascending = False)
```

```python
data = pd.read_csv("Pfizer_1.csv")
```

```python
data
```

```python
# Melt takes the columns that you do not want to melt
pd.melt(data, id_vars = ["Date", "Parameter", "Drug_Name"])
```

```python
data_melt = pd.melt(
    data,
    id_vars = ["Date", "Parameter", "Drug_Name"],
    var_name = "Time",
    value_name = "Reading"
)
```

```python
data_melt.head()
```

```python
data_melt.pivot(
    index = ["Date", "Drug_Name", "Parameter"],
    columns = "Time",
    values = "Reading"
)
```

```python
data_melt.pivot(
    index = ["Date", "Drug_Name", "Parameter"],
    columns = "Time",
    values = "Reading"
).reset_index()
```

```python
data_melt # [Date, Drug_Name, Time, Pressure, Temperature]
```

```python
data_tidy = data_melt.pivot(
    index = ["Date", "Time", "Drug_Name"],
    columns = "Parameter",
    values = "Reading"
).reset_index()
```
