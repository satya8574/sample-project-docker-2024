```python
sales_data = data[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]]
```

```python
sales_data.T
```

```python
region_sales = sales_data.T.sum(axis = "columns")
```

```python
plt.pie(
    region_sales,
    labels = region_sales.index,
)
plt.show()
```

```python
plt.pie(
    region_sales,
    labels = region_sales.index,
    startangle = 90,
    explode = (0.2, 0, 0, 0),
    autopct = "%.2f%%"
)
plt.show()
```
