```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('vgsales.csv')
data.tail(20)
```

```python
# How can we plot the sales of the longest running game over the years?
```

```python
data["Name"].value_counts()
```

```python
ice_hockey = data.loc[data["Name"] == "Ice Hockey"]
```

```python
ice_hockey.head()
```

```python
sns.lineplot(x = "Year", y = "NA_Sales", data = ice_hockey)
plt.show()
```

```python
# What if I want to change the axis and chart titles?

sns.lineplot(x = "Year", y = "NA_Sales", data = ice_hockey)
plt.xlabel("YEAR")
plt.ylabel("SALES")
plt.title("Ice Hockey Sales Trend")
plt.show()
```

```python
# What if we want to change the color here?

sns.lineplot(x = "Year", y = "NA_Sales", data = ice_hockey, color = "r")
plt.xlabel("YEAR")
plt.ylabel("SALES")
plt.title("Ice Hockey Sales Trend")
plt.show()

# https://matplotlib.org/2.0.2/api/colors_api.html
```

```python
# How can we limit our plot to only include data from 1995 to 2010?

# Formatting
plt.title("Ice Hockey Sales Trend")
plt.xlabel("YEAR")
plt.ylabel("SALES")
plt.xlim(left = 1995, right = 2010)

# Plot
sns.lineplot(x = "Year", y = "NA_Sales", data = ice_hockey)

# Show
plt.show()
```

```python
baseball = data.loc[data["Name"] == "Baseball"]
sns.lineplot(x = "Year", y = "NA_Sales", data = baseball)
```

```python
# How can we plot multiple plots together?

sns.lineplot(x = "Year", y = "NA_Sales", data = ice_hockey)
sns.lineplot(x = "Year", y = "NA_Sales", data = baseball)
plt.show()
```

```python
# How to I get a legend here?

sns.lineplot(x = "Year", y = "NA_Sales", data = ice_hockey, label = "Ice Hockey")
sns.lineplot(x = "Year", y = "NA_Sales", data = baseball, label = "Baseball")
plt.show()
```

```python
sns.lineplot(x = "Year", y = "NA_Sales", data = ice_hockey)
sns.lineplot(x = "Year", y = "NA_Sales", data = baseball)
plt.legend(["Baseball", "Ice Hockey"]) # Order of the plots needs to be remembered
plt.show()

# NOT RECOMMENDED ^
```

```python
# How can I move around the legend?

sns.lineplot(x = "Year", y = "NA_Sales", data = ice_hockey, label = "Ice Hockey")
sns.lineplot(x = "Year", y = "NA_Sales", data = baseball, label = "Baseball")
plt.legend(loc = (-0.5, 0.5))
plt.show()
```

```python
sns.lineplot(x = "Year", y = "NA_Sales", data = ice_hockey, label = "Ice Hockey")
sns.lineplot(x = "Year", y = "NA_Sales", data = baseball, label = "Baseball")
plt.legend(loc = "upper left")
plt.show()
```

```python
sns.lineplot(x = "Year", y = "NA_Sales", data = ice_hockey, label = "Ice Hockey")
sns.lineplot(x = "Year", y = "NA_Sales", data = baseball, label = "Baseball")
plt.legend(loc = "lower left", ncol = 2, frameon = False)
plt.show()
```

```python
sns.lineplot(x = "Year", y = "NA_Sales", data = ice_hockey, label = "Ice Hockey")
sns.lineplot(x = "Year", y = "NA_Sales", data = baseball, label = "Baseball")
plt.legend(loc = "lower left", ncol = 2, frameon = False)
plt.grid()
plt.show()

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html
```

```python
# Trying to obserbe correlation in Rank and Global Sales

sns.lineplot(data = data, x = "Rank", y = "Global_Sales")
```

```python
sns.scatterplot(data = data, x = "Rank", y = "Global_Sales")
```

```python
top3_pub = data["Publisher"].value_counts().index[:3]
top3_genres = data["Genre"].value_counts().index[:3]
top3_platforms = data["Platform"].value_counts().index[:3]
```

```python
top3_data = data.loc[(data["Publisher"].isin(top3_pub)) & (data["Genre"].isin(top3_genres)) & (data["Platform"].isin(top3_platforms))]
```

```python
top3_data.head(10)
```

```python
sns.countplot(x = "Publisher", hue = "Platform", data = top3_data)
plt.show()
```

```python
sns.boxplot(x = "Publisher", y = "Global_Sales", data = top3_data)
plt.show()
```

```python
sns.barplot(data = top3_data, x = "Genre", y = "Global_Sales", estimator = np.mean)
```

```python
sns.scatterplot(x = top3_data["NA_Sales"], y = top3_data["Global_Sales"])
```

```python
fig = plt.figure(figsize = (10, 10))

plt.subplot(2, 2, 1) # num_rows, num_columns, position
sns.scatterplot(x = "NA_Sales", y = "Global_Sales", data = top3_data)

plt.subplot(2, 2, 2)
sns.scatterplot(x = "EU_Sales", y = "Global_Sales", data = top3_data, color = "red")

plt.subplot(2, 2, 3)
sns.scatterplot(x = "JP_Sales", y = "Global_Sales", data = top3_data, color = "orange")

plt.subplot(2, 2, 4)
sns.scatterplot(x = "Other_Sales", y = "Global_Sales", data = top3_data, color = "green")

fig.suptitle("Sales Across Regions vs Global Sales")
```

```python
fig = plt.figure(figsize = (10, 10))

plt.subplot(2, 3, 1) # num_rows, num_columns, position
sns.scatterplot(x = "NA_Sales", y = "Global_Sales", data = top3_data)

plt.subplot(2, 3, 3)
sns.scatterplot(x = "EU_Sales", y = "Global_Sales", data = top3_data, color = "red")

plt.subplot(2, 3, 4)
sns.scatterplot(x = "JP_Sales", y = "Global_Sales", data = top3_data, color = "orange")

plt.subplot(2, 3, 6)
sns.scatterplot(x = "Other_Sales", y = "Global_Sales", data = top3_data, color = "green")

fig.suptitle("Sales Across Regions vs Global Sales")
```
