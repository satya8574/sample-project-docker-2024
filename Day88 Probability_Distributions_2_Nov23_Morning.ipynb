```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Roll a die 100000 times and plot the observations as a histogram
data = np.random.uniform(1, 7, 100000)

# Any number between 1 and 2 -> 1 Bucket
# Any number between 2 and 3 -> 2 Bucket...
```

```python
data[:20]
```

```python
plt.hist(data, bins = 6, edgecolor = "black", align = "mid", rwidth = 0.8, color = "skyblue", density = True)
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("Histogram of a Uniform Distribution (Die Roll Simulation)")
plt.show()
```
